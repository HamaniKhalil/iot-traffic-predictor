# Import

```python
import pandas as pd
import numpy as np

# encodage
from sklearn.preprocessing import minmax_scale
from Encoder import Encoder
```

# Lecture du dataset pré-traité

```python
preprocessed_dataset_file = "dataset/preprocessed/main-traces-preprocessed.csv"
```

```python
traffic_df = pd.read_csv(preprocessed_dataset_file)
traffic_df.head()
```

# Encodage des colonnes

## SourceID

```python
source_id_encoder = Encoder.Encoder()
source_id = traffic_df["sourceID"].tolist()
source_id_encoded = source_id_encoder.encode_data(source_id)
```

## Source Address

```python
source_address_encoder = Encoder.Encoder()
source_address = traffic_df["sourceAddress"].tolist()
source_address_encoded = source_id_encoder.encode_data(source_address)
```

## Source Type

```python
source_type_encoder = Encoder.Encoder()
source_type = traffic_df["sourceType"].tolist()
source_type_encoded = source_id_encoder.encode_data(source_type)
```

## Source Location

```python
source_location_encoder = Encoder.Encoder()
source_location = traffic_df["sourceLocation"].tolist()
source_location_encoded = source_id_encoder.encode_data(source_location)
```

## Destination Service Address

```python
destination_service_addr_encoder = Encoder.Encoder()
destination_service_addr = traffic_df["destinationServiceAddress"].tolist()
destination_service_addr_encoded = source_id_encoder.encode_data(destination_service_addr)
```

## Destination Service Type

```python
destination_service_type_encoder = Encoder.Encoder()
destination_service_type = traffic_df["destinationServiceType"].tolist()
destination_service_type_encoded = source_id_encoder.encode_data(destination_service_type)
```

## Destination Location

```python
destination_location_encoder = Encoder.Encoder()
destination_location = traffic_df["destinationLocation"].tolist()
destination_location_encoded = source_id_encoder.encode_data(destination_location)
```

## Accessed Node Address

```python
accessed_node_addr_encoder = Encoder.Encoder()
accessed_node_addr = traffic_df["accessedNodeAddress"].tolist()
accessed_node_addr_encoded = source_id_encoder.encode_data(accessed_node_addr)
```

## Accessed Node Type

```python
accessed_node_type_encoder = Encoder.Encoder()
accessed_node_type = traffic_df["accessedNodeType"].tolist()
accessed_node_type_encoded = source_id_encoder.encode_data(accessed_node_type)
```

## Operation

```python
operation_encoder = Encoder.Encoder()
operation = traffic_df["operation"].tolist()
operation_encoded = source_id_encoder.encode_data(operation)
```

## Value

```python
value_list = traffic_df["value"]
```

```python
no_num_dict = {}
```

```python
value_code = -1
for cursor in value_list:
    try:
        float(cursor)
    except ValueError:
        if cursor not in no_num_dict:
            no_num_dict[cursor] = value_code
            value_code -= 1
```

```python
no_num_dict
```

```python
value_encoded = []
```

```python
for cursor in value_list:
    try:
        value_encoded.append(float(cursor))
    except ValueError:
        if cursor in no_num_dict:
            value_encoded.append(no_num_dict[cursor])
```

## Timestamp

```python
timestamp = traffic_df["timestamp"].tolist()
timestamp_encoded = minmax_scale(X=timestamp)
```

## Normality

```python
normality = traffic_df["normality"]
```

```python
normality_values = {}
```

```python
normality_code = 0
for cursor in normality:
    if cursor not in normality_values:
        normality_values[cursor] = normality_code
        normality_code += 1
```

```python
normality_values
```

```python
normality_encoded = []
```

```python
for cursor in normality:
    normality_encoded.append(normality_values[cursor])
```

```python
normality_encoded
```

# Sauvegarde du dataframe encodé

```python
traffic_encoded_df = pd.DataFrame({
    "sourceID" : source_id_encoded,
    "sourceAddress" : source_address_encoded,
    "sourceType" : source_type_encoded,
    "sourceLocation" : source_location_encoded,
    "destinationServiceAddress" : destination_service_addr_encoded,
    "destinationServiceType" : destination_service_type_encoded,
    "destinationLocation" : destination_location_encoded,
    "accessedNodeAddress" : accessed_node_addr_encoded,
    "accessedNodeType" : accessed_node_type_encoded,
    "operation" : operation_encoded,
    "value" : value_encoded,
    "timestamp" : timestamp_encoded,
    "normality" : normality_encoded
})
```

```python
encoded_dataset_file = "dataset/encoded/main-traces-encoded.csv"
```

```python
traffic_encoded_df.to_csv(encoded_dataset_file, index=False)
```
